```python
# Siyah rengi = [0,0,0]
# Kırmızı rengi = [255,0,0]
# Turuncu-sarı = [253,222,2]

height, width = 150, 210

flag_germany = np.zeros((height,width,channel),dtype=np.uint8)

flag_germany[0:50,:,:] = [0,0,0] # görüntünün üst tarafını siyah yaptık
flag_germany[50:100,:,:] = [255,0,0] # görüntünün orta tarafını kırmızı yaptık
flag_germany[100:,:,:] = [253,222,2] # görüntünün alt tarafını turuncu yaptık
```

```python
plt.imshow(flag_germany)
plt.axis("off")
plt.title("Germany Flag")
```

```python
# height ve width boyutlarında fenerbahçe renklerini dikey bir şekilde oluşturun

flag_fenerbahce = np.zeros((height,width,channel),dtype=np.uint8)

flag_fenerbahce[:,0:105,:] = [255,255,0] # görüntünün ilk kısmını sarı yaptık (dikeyde)
flag_fenerbahce[:,105:,:] = [0,0,255] # görüntünün ikinci kısmını mavi yaptık (dikeyde)
```

```python
plt.imshow(flag_fenerbahce)
plt.axis("off")
plt.title("ŞAMPİYON FENERBAHÇE")
```

```python
# Fenerbahçe amblemini GS ambleminin renklerine dönüştürelim.
# Bunun için ilk olarak FB amblemindeki mavi renklerini seçmemiz lazım
# Mavi rengi için RGB kodları = [0,0,255]
# Biz bu değerleri görüntüde yer alan bütün satır ve bütün sütunlarda arayacağız
# NOT: bu işlemi yaparken 3 kanal için de tek tek arama yapacağız

renk_koordinat_mavi = (flag_fenerbahce[:,:,0]==0) & (flag_fenerbahce[:,:,1]==0) & (flag_fenerbahce[:,:,2]==255)
```

```python
# yukarıda, mavi renklerini tuttuğumuz koordinatı tekrar flag_fenerbahce
# değişkeni içerisine verip, bu koordinatlardaki yeni renkleri belirliyoruz

flag_fenerbahce[renk_koordinat_mavi] = [255,0,0]
```

```python
plt.imshow(flag_fenerbahce)
plt.axis("off")
plt.title("Dünyanın En Şanslı Takımı")
```

```python
# Bu defa sarı rengini beyaz, kırmızı rengini de siyah rengiyle değiştirin

# sarı renginin koordinatlarını belirleme
renk_koordinat_sari = (flag_fenerbahce[:,:,0]==255) & (flag_fenerbahce[:,:,1]==255) & (flag_fenerbahce[:,:,2]==0)

# kırmızı renginin koordinatlarını belirleme
renk_koordinat_kirmizi = (flag_fenerbahce[:,:,0]==255) & (flag_fenerbahce[:,:,1]==0) & (flag_fenerbahce[:,:,2]==0)
```

```python
# sarı rengini beyaza dönüştürme
flag_fenerbahce[renk_koordinat_sari] = [255,255,255]

#kırmızı rengini siyaha dönüştürme
flag_fenerbahce[renk_koordinat_kirmizi] = [0,0,0]
```

```python
fig = plt.figure(figsize=(6,4),facecolor="purple")
plt.imshow(flag_fenerbahce)
plt.axis("off")
plt.title("Gariban Takım")
```
