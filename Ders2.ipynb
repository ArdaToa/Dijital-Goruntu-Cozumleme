```python
# FLIP forks younun islevine manuel olarake tanmlanak
my_flip_img_1=my_rgb_img[:,::-1,:]#dikey eksen tersine döner
my_flip_img_2=my_rgb_img[::-1,:,:]#yatay eksen tersine

my_flip_img_3=my_rgb_img[::-1,::-1,:]
```

```python
#kendi fonsiyonumuz
def my_flip(org_img, number):
    if number==0:
        org_img= org_img[::-1,:,:]
    elif number==1:
        org_img=org_img[:,::-1,:]
    elif number==-1:
        org_img=org_img[::-1,::-1,:]
    else:raise ValueError("Lütfen doğru parametrelerigirdiğinizden emin olun!")
    return org_img

my_new_vertical=my_flip(my_rgb_img,0)
```

```python
my_new_vertical =my_flip(my_rgb_img,0)
my_new_horizontal=my_flip(my_rgb_img,1)
my_new_both= my_flip(my_rgb_img,-1)
```

```python
fig=plt.figure(figsize=(10,7))

plt.subplot(2,2,1)
plt.imshow(my_rgb_img)
plt.axis("off")
plt.title("Original Image")

plt.subplot(2,2,2)
plt.imshow(my_new_vertical)

plt.subplot(2,2,3)
plt.imshow(my_new_horizontal)
plt.axis("off")
plt.title("Horizontal Image")

plt.subplot(2,2,4)
plt.imshow(my_new_horizontal)
plt.axis("off")
plt.title("Horizontal Image")
```
