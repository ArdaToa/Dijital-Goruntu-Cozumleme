```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
"""
Görüntü işlemde morfolojik işlem, görüntülerdeki yapısal öğeleri inceleyen ve manipüle eden işlemler grubuna verilen addır. 
Bu işlemler genellikle ikili (binary) veya gri tonlamalı görüntüler üzerinde gerçekleştirilir ve şekil, boyut, yapı gibi özellikleri 
çıkarmaya veya değiştirmeye yönelik matematiksel tekniklerdir. Morfolojik işlemler, özellikle nesne sınırlarını, şekil özelliklerini ve bağlantıları 
analiz etme ve iyileştirme amacıyla kullanılır.
"""
```

```python
my_binary_img = cv2.imread("./paint.png", 0)
print(my_binary_img.shape)
```

```python
my_kernel = np.ones((5, 5), dtype=np.uint8)
```

```python
my_erosion_img = cv2.erode(my_binary_img, my_kernel, iterations=2)
```

```python
fig = plt.figure(figsize=(10, 7))

plt.subplot(1,2,1)
plt.imshow(my_binary_img, cmap="gray")
plt.axis("off")
plt.title("Original Image")


plt.subplot(1,2,2)
plt.imshow(my_erosion_img, cmap="gray")
plt.axis("off")
plt.title("Eresion Image")
```

```python
my_dilation_img = cv2.dilate(my_binary_img, my_kernel, iterations=6)
```

```python
fig = plt.figure(figsize=(10, 7))

plt.subplot(1,3,1)
plt.imshow(my_binary_img, cmap="gray")
plt.axis("off")
plt.title("Original Image")


plt.subplot(1,3,2)
plt.imshow(my_erosion_img, cmap="gray")
plt.axis("off")
plt.title("Eresion Image")

plt.subplot(1,3,3)
plt.imshow(my_dilation_img, cmap="gray")
plt.axis("off")
plt.title("Dilate Image")
```

```python
new_img=cv2.imread("./bst.png", 0)
```

```python
new_img.shape
```

```python
white_noise = np.random.randint(0,2,size=new_img.shape[:2])
white_noise *= 255
```

```python
plt.imshow(white_noise, cmap ="gray")
plt.axis("off")
plt.title("Full Noised Image")
```

```python
my_noised_img = white_noise + new_img
plt.imshow(my_noised_img, cmap = "gray")
plt.axis("off")
plt.title("Noised Image")
```

```python
my_opening_img= cv2.morphologyEx(my_noised_img.astype(np.float32),
                                cv2.MORPH_OPEN, my_kernel)
```

```python
fig = plt.figure(figsize=(10, 7))

plt.subplot(1,3,1)
plt.imshow(new_img, cmap="gray")
plt.axis("off")
plt.title("Original Image")


plt.subplot(1,3,2)
plt.imshow(my_noised_img, cmap="gray")
plt.axis("off")
plt.title("Image with Noise")

plt.subplot(1,3,3)
plt.imshow(my_opening_img, cmap="gray")
plt.axis("off")
plt.title("Opening Image")
```

```python
black_noise = np.random.randint(0,2,size=new_img.shape[0:2])
black_noise *= -255
```

```python
my_black_noise_img = black_noise + new_img
```

```python
my_black_noise_img[my_black_noise_img <= -245] = 0
```

```python
plt.imshow(my_black_noise_img,cmap="gray")
plt.axis("off")
plt.title("Black Noised Image")
```

```python
my_closing_img= cv2.morphologyEx(my_black_noise_img.astype(np.float32),
                                cv2.MORPH_CLOSE, my_kernel)
```

```python
fig = plt.figure(figsize=(10, 7))

plt.subplot(1,3,1)
plt.imshow(new_img, cmap="gray")
plt.axis("off")
plt.title("Original Image")


plt.subplot(1,3,2)
plt.imshow(my_black_noise_img, cmap="gray")
plt.axis("off")
plt.title("Image with Black Noise")

plt.subplot(1,3,3)
plt.imshow(my_closing_img, cmap="gray")
plt.axis("off")
plt.title("Closing Image")
```

```python
my_grandient_1 = cv2.morphologyEx(my_binary_img, cv2.MORPH_GRADIENT, my_kernel)
my_grandient_2 = cv2.morphologyEx(new_img, cv2.MORPH_GRADIENT, my_kernel)
```

```python
fig = plt.figure(figsize=(10, 7))

plt.subplot(2,2,1)
plt.imshow(my_binary_img, cmap="gray")
plt.axis("off")
plt.title("Original Image")


plt.subplot(2,2,2)
plt.imshow(my_grandient_1, cmap="gray")
plt.axis("off")
plt.title("Gradient 1")

plt.subplot(2,2,3)
plt.imshow(new_img, cmap="gray")
plt.axis("off")
plt.title("Original Image")

plt.subplot(2,2,4)
plt.imshow(my_grandient_2, cmap="gray")
plt.axis("off")
plt.title("Gradient 2")
```
