```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Görüntüyü, renkli olsa bile her zaman gri seviyede okumalıyız.
# ya, bgr olarak okuyup sonradan gray seviyesine çevirirsiniz
# ya da direkt başlangıçta gray seviyede okursunuz

my_bgr_img =cv2.imread("./bonus.jpg")
my_img = cv2.imread("./bonus.jpg",0)

# Daha sonra görüntüye eşikleme işlemi uyguluyoruz. Eşikleme işlemiyle
# görüntünün zaten siyah-beyaz seviyede ayrımı ortaya çıkıyor.
```

```python
ret, thresh = cv2.threshold(my_img, 127, 255, cv2.THRESH_BINARY)
# görüntüdeki kontuları bulma
```

```python
countur, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# yukarıdaki kontur bulma fonksiyonundan 2 değer döner.
# ilk değer, bulunan kontorların listesini içerir
# ikinci değer bu konturlar arasındaki ebeveyn ilişkisini gösterir
# bize lazım olan, göstereceğimiz asıl değer konturların listesini saklayan
# ilk değişkendir.

# Elde ettiğimiz konturları, görüntü üzerinde birleştirip çizdiriyoruz
# renkleri ayırt edilebilsin diye yeşil renkle (0,255,0) çizdiriyoruz
```

```python
cv2.drawContours(my_bgr_img, countur, -1, (0, 255, 0), 2)

# fonksiyondaki parametreler:
# -1 : bütün  konturları göster, satır-sütun tamamı olacak şekilde
# (0,255,0) RGB'nin kodlarıdır. Bu renk kodu yeşil rengini temsil eder
# konturu çizdirirken yeşil renkle çizdirecek
# 2 değeri de çizginin (yeşil renkle çizdirilecek kontur çizgisinin)
# kalınlığıdır

cv2.imshow("Original Countured Image",my_bgr_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
