```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#Pythonda path kullanımı :
#./ ifadesi, kod dosyanızın çalıştığı klasör karşılar.
#Örneğin, eğer kodunuz C:\Users\Arda dizini altındaysa, siz 
#her defasında "C:\\Users\\Arda" olarak aratırsınız ya da 
#r"C:\Users\Arda" bu şekilde bir path yolu verirsiniz.
#Ancak, programlama dillerdinde özellikle pythonda ./ ifadesi ile
#o an aktif oalrak çalıştığınız klasör yolunu ifade edersiniz, uzun uzadıya
#her deafında parthi sıfırdan 
my_img = cv2.imread("./Lion_waiting_in_Namibia.jpg")
```

```python
print(type(my_img))
# buraasının çıktısı numpy.ndarray olmalı (çünkübütün görüntüler bu tiptedir.)
```

```python
print(my_img.shape)
print(f"Görüntünün yüksekliği (satır sayısı) = {my_img.shape[0]}")
print(f"Görüntünün genişliği (sütün sayısı) = {my_img.shape[1]}")
```

```python
#BGR olarak renkli bir şekilde okunana görünttüyü, Gray seviyesinde 
#yani tek kanallı bir hale dönüştürüyoruz.
my_gray_img = cv2.cvtColor(my_img, cv2.COLOR_BGR2GRAY)
print(my_gray_img.shape)
```

```python
#Thresholding, siyah-beyaz yani gri seviyede tek kannalı görüntülerin 
#kenar tespiti, nesne tanıma, nesne ayırt etme, tıbbi alanda
#hücre görüntülerinin maskelenmesi (ground truth mask images) gibi alanlarda
#kullanılan klasık bir görüntü işleme tekniğidir.

#Bir eşik değeri seçilir, siyah-beyaz gri görüntünün her bir pikseli
#bu eşik değeri ile karşılaştırılır. Eğer ilgili piksel değeri, bu eşik değerinden
#düşükse yeni bir piksel değeri alır, yüksekse yine yeni bir piksel değeri alır. Farklı 
#eşikleme yöntemlerinin olmamasının nedeni, bu yeni piksel değerinin kaç olacağı ile ilgilidir.
```

```python
_, my_binary_thresh_img = cv2.threshold(my_gray_img, thresh = 125,
                                       maxval = 225, type = cv2.THRESH_BINARY)
```

```python
print(f"Orijinal Piksel Değeri = {my_gray_img[500, 900]}")
print(f"Binary thresh değeri = {my_binary_thresh_img[500, 900]}")
```

```python
fig = plt.figure(figsize = (10, 7))

plt.subplot(1,2,1)
plt.imshow(my_gray_img, cmap = "gray")
plt.axis("off")
plt.title("Original Gray Image")

plt.subplot(1,2,2)
plt.imshow(my_binary_thresh_img, cmap = "gray")
plt.axis("off")
plt.title("Binary Thresh Image")
```

```python
_, my_binary_inv_thresh_img = cv2.threshold(my_gray_img, thresh = 125,
                                       maxval = 225, type = cv2.THRESH_BINARY_INV)
```

```python
#Eşikleme yönteminin sayısal olarak ispatı
print(f"Orijinal Piksel Değeri = {my_gray_img[500, 900]}")
print(f"Binary thresh değeri = {my_binary_thresh_img[500, 900]}")
print(f"Binary inverse thresh değeri = {my_binary_inv_thresh_img[500, 900]}")
```

```python
fig = plt.figure(figsize = (10, 7))

plt.subplot(1,3,1)
plt.imshow(my_gray_img, cmap = "gray")
plt.axis("off")
plt.title("Original Gray Image")

plt.subplot(1,3,2)
plt.imshow(my_binary_thresh_img, cmap = "gray")
plt.axis("off")
plt.title("Binary Thresh Image")

plt.subplot(1,3,3)
plt.imshow(my_binary_inv_thresh_img, cmap = "gray")
plt.axis("off")
plt.title("Binary İnverse Thresh Image")
```

```python
#cv2.resize(original_image ,(new_width, new_height),cv2.INTER_LANCZOS4)
```

```python
def smaller_image(img_org):
    new_height = img_org.shape[0] - int(img_org.shape[0]*0.25)
    new_weight = img_org.shape[1] - int(img_org.shape[1]*0.25)

    new_img = cv2.resize(img_org ,(new_weight, new_height),cv2.INTER_LANCZOS4)

    return new_img

def larger_image(img_org):
    new_height = img_org.shape[0] + int(img_org.shape[0]*0.25)
    new_weight = img_org.shape[1] + int(img_org.shape[1]*0.25)
    
    new_img = cv2.resize(img_org,(new_weight, new_height),cv2.INTER_LANCZOS4)

    return new_img
```

```python
my_larger_img = larger_image(my_gray_img)
my_smaller_img = smaller_image(my_gray_img)
```

```python
print(f"Orijinal Boyut = {my_gray_img.shape}")
print(f"Büyütülmüş Boyut = {my_larger_img.shape}")
print(f"Küçültülmüş Boyut = {my_smaller_img.shape}")
```
